**Ejercicio 1**

```python
import math as m #en esta paqueter√≠a encontramos pi
```

```python
def my_cilinder(r, h): #ùëü y ‚Ñé son el radio y la altura del cilindro
    
    """This function returns you the superficial area and the volume in a list of a cilinder"""
    
    s = []
    v = [] 
    
    area = (m.pi)*(r)**2 #aqu√≠ estamos calculando el √°rea del c√≠rculo 
    s.append(area) #estamos agregando el valor de √°rea a una lista 
    vol = area*h #calculamos el volumen
    v.append(vol) #agrego el volumen a una lista 
    resul = s + v #concatenamos la listas 
    print(resul) #imprimo la lista con el √°rea y el volumen respectivamente
    
```

```python
my_cilinder(2, 6)
```

**Ejercicio 2**

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
def my_poligon(n): #n es el n√∫mero de lados que tendr√° el pol√≠gono 
    
    """This function graficates a poligon of n sides with polar coordinates"""
    
    r = []
    a = []
    
    for i  in range (0, n):
        radios = 1
        r.append(radios) #creo mi lista de todos los radios
        ang = i/n *2* np.pi 
        a.append(ang) #creo mi lista de todos los √°ngulos
    
    r+= r[:1] #le agrego el √∫ltimo valor para que cierre el pol√≠gono
    a+= a[:1] #le agrego el √∫ltimo valor para que cierre el pol√≠gono
    
    print(r, a)
    
    figure(figsize = (10, 30), dpi = 60)
    plt.polar(a, r, c = "seagreen")
    plt.title(f'Un pol√≠gono de {n} lados')
    
    
```

```python
my_poligon(8)
```

```python
doc = open('Data-A49.dat', 'r') #abro el archivo en modo lectura
lineas = doc.readlines()
# quitar la primera l√≠nea , que tiene los
# nombres de las columnas
# y tambi√©n la √∫ltima l√≠nea que est√° vac√≠a 
lineas = lineas[1:-1] 
doc.close()

x, y, sd = [], [], []

for linea in lineas: # linea es str
    linea = linea.split() # linea ahora es [str, str, str]
    # elemento en x es linea[0]
    # elemento en y es linea[1]
    # elemento en sd es linea[2]

    x.append(float(linea[0]))
    y.append(float(linea[1]))
    sd.append(float(linea[2]))
    
print(x)
print(y)
print(sd)

plt.figure(figsize=(7,7))
plt.plot(x,y, c = "violet", marker = ".") 
plt.ylabel('Y', fontsize = 13, rotation = 0)
plt.xlabel('X', fontsize = 13)
plt.xlim(0, 900)
plt.ylim(-400,500)
plt.title("Gr√°fica de valores X y Y")
caption ="Gr√°fica del archivo Data-A49"
plt.figtext(0.5, -0.0009, caption, wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
```
