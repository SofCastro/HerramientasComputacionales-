## 2.


Escriba una función que se llame `A_sin()` que reciba como parámetro a `t` y argumentos `A = 1`, `omega = 2*np.pi` y `phi = np.pi/2` y que regrese $f(t) = A\sin(t\omega_0 + \phi )$. Nota: La función debe conntar con un *docstring*. En las celdas a continucación se muestra el funcionamiento de `A_sin()`

```python
def A_sin(t, A = 1, omega = 2*np.pi, phi = np.pi/2):
    
    ''' Esta función nos determina el valor de seno multiplicado por A de t por omega más phi '''
    
    x = (t*omega) + phi  #multiplico t por omega y le sumo phi para que sea el argumento de seno  
    y = A*np.sin(x)  #del producto anterior le multiplico 
    
    return y #nos regresa el producto ya calculado 
```

Una función que cuenta con infinitas derivadas y además estás son continuas, se puede escribir como la siguiente serie alrededor de un punto $x_0$ del dominio de la función:

$$f(x) = \displaystyle \sum_{k = 0}^{\infty}\dfrac{f^{k}(x_0)}{k!}(x - x_0)^k$$

Donde $f^{n}(x_0)$ denota la n-ésima derivada de la función $f$ evaluada en $x_0$. Sin embargo podemos aproximar a la función $f$ truncando la serie a un cierto orden $n$ y lo que nos queda es un polinomio de orden $n$, a continuación se muestran algunas aproximaciones de ciertas funciones:

* $\exp(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{x^k}{k!}$

* $\ln(1- x)  \approx - \displaystyle \sum_{k = 0}^{n}\dfrac{x^k}{k}$

* $\sin(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{(-1)^k}{(2k +1)!}x^{2k+1}$


* $\cos(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{(-1)^k}{(2k )!}x^{2k}$

Implemente cuatro funciones cuyos nombres sean: `my_exp()`, `my_ln()`, `my_sin()` y `my_cos()` que aproximen las respectivas funciones a un orden $n$ alrededor de un punto $x_0$. Por ejemplo:

```python
def factorial(n): #definimos la función factorial
    
    """ La función calcula el factorial de un número  """
    
    i = 1 #iniciamos la función
    count = 1 #establecemos nuestro contador 
    while i < n: #para que se cumpla el ciclo, i tiene que ser menos a n
        i+= 1 #se le suma a i 1
        count*= i #el contador se le mutiplica i 
    return count #se regresa el contador 
```

```python
factorial(5) #corremos la función
```

```python
?factorial
```

```python
def my_sin(x, n = 5):
    
    """ La función seno recibe un punto x, y un número límite que será n """
    k = 0 #k empieza en cero
    count = 0 #establecemos nuestro contador
    while k < n:
        count = count + (((-1)**k)*(x**(2*k + 1)))/factorial(2*k +1) #estos son los cálculos de la fórmula que se le sumará al contador
        k+= 1 #k aumenta 1
    return count #se regresa el contador
```

```python
my_sin(np.pi, 3)
```

```python
def my_exp(x, n = 2):
    """ La función exponencial recibe un punto x y también n """
    k = 0  #k empieza en cero por que asi fue definida en la formula 
    aux = 0 #utilizamos la variable aux que nos sirve como un acumulador
    while k < n: #k tiene que ser menor a n
        aux = aux + (x**k)/factorial(k)  #aqui va guardando los valores que va adquiriendo la función
        k+= 1 #k va a umentar de 1 en 1 
    return aux #regresa el contador
```

```python
def my_ln(x, n = 8):  #Definimos la función 
    
    """ La función my_ln recibe un punto x que se usará en todo el ciclo, mientras que n es hasta dónde llegará """
    
    k = 1
    aux = 0  #inicializamos el valor de aux en 0
    while k < n:  #empleamos un ciclo while 
        aux = aux + ((-1)*(x**k))/k   #aux nos está calculando el valor de aux más el procedimiento empleado
        k+= 1  #para que el ciclo avance se le aumentara el valor de uno a k 
    return aux #para finalizar nos devvuelve el valor de aux 
```

```python
def my_cos(x, n = 5): #Definimos la función  
    
    """ La función coseno recibe un punto x que será el mismo durante todo el ciclo y otro valor n que será el total de sumas a hacer  """
    
    k = 0 #Definimos el valor de k en 0 
    aux = 0 #inicializamos a aux con el valor 0
    while k < n:
        aux = aux + (((-1)**k)*(x**(2*k)))/factorial(2*k)  #se realiza la fórmula que se le irá sumando a aux
        k+=1
    return aux  #regresamos el valor aux
```

**¿Que puede concluir de este tipo de aproximación?**

Conforme la función se va acercando a 0, tomará valores dependiendo la función. La función exponencial nunca llega a 0, pero si nos aproximamos, podemos decir que va haciéndose como una recta; la función ln está indefinida en 0, pero al irnos acercando toma valores negativos; coseno al aproximarse a 0 nos da 1 y finalmente seno al aproximarse a 0 nos dará 0.
