# 1. El Cifrado de Julio César

Durante sus campañas militares, el general romano Julio César escribía mensajes a sus subordinados de la siguiente forma:

Sustituía cada letra del abecedario por la letra situada tres lugares más allá en el orden alfabético. Es decir, la letra A se convertiría en D; la B, en E; la C en F y así sucesivamente. Y del mismo modo para la cola: la X sería una A; la Y una B y la Z una C. En resumen, las letras se desplazan tres lugares en el orden alfabético.

Por ejemplo, en lugar de escribir CENTURIA escribía FHPWXULD.

# 3. Medidas de tendencia central  y dispersión

Las medidas de tendencia central, se refire a las medidas estadisticas que permiten resumir en un solo valor a un conjunto de  valores. Es más facil calcular estas medidas cuando los datos se encuentran ordenado en una tabla.

#### ¿Cómo calcular medidas de tendencia central en ` phyton ` ? 
Se puede utilizar la libreria ` statistics ` ya que  agrupa un conjunto de funciones para cálculo estadístico. 

Para importar` statistics `:

```python
import statistics as stats
```

| Función   |   Descripción 
|-----------|------------------
|  mean()   |    Calcula la media o promedio de los datos        
|  median()         | Calcula la mediana de los datos              
|   mode()        |  Calcula la moda de los datos  
|   pvariance()        |  Calcula  la varianza de una población     
|   variance()        |  Calcula la varianza de una muestra  
|    stedev()   | Calcula la desviación estandar de una muestra
|   pstdev()  |Calcula la desviación estandar de una población

**Media o promedio:** Es la suma de los valores del grupo de datos dividida entre la cantidad de valores. Su fórmula se puede describir de la siguiente manera: $$Media = \frac{suma\ de\ los\ valores}{cantidad\ de\ valores}$$

$$ \bar{x} = \frac{\sum x}{n}$$

```python
qualification = [9, 10, 8, 7, 6, 10, 4, 6, 9, 7, 8, 6, 9, 5, 6 ] #conjunto de datos 
print(stats.mean(qualification)) # se imprime el valor de la media o promedio implementada  por la libreria statitics 
```

**Mediana:** Hace referencia al valor del elemento central de un conjunto. 
Para poder encontrar el valor de la mediana es necesario ordenar los elementos del conjunto del valor menor al mayor, y despues encontrar el valor de la posición central. 
$$M_{e} = L_{i} + \frac{\frac{n}{2}-F_{i-1}}{f_{i}}*a_{i}$$
*Donde*: $L_{i}: Límite\ inferior\ de\ la\ clase\ donde\ se\ encuentra\ la\ mediana$\
$\frac{n}{2}: Semisuma\ de\ las\ frecuencias\ absolutas$\
$f_{i}:Frecuencia\ absulta$\
$F_{i-1}:Frecuencia\ acumulada\ anterior\ a\ la\ clase\ mediana$\
$a_{i}:Amplitud\ de\ la\ clase$


```python
qualification  = [9, 10, 8, 7, 6, 10, 4, 6, 9, 7, 8, 6, 9, 5, 6 ] #conjunto de datos 
print(stats.median(qualification )) # se imprime el valor de la mediana implementada  por la libreria statitics 
```

**Modo:** También denominada moda o promedio típico de un conjunto de valores, hace referencia al valor que se repita má veces en un conjunto.
Tambíen nos indica que es el valor más probable a ser seleccionado si se escoge un valor al azar del conjunto.  
$$M_{o}=L_{i}+\frac{f_{i}-f_{i-1}}{(f_{i}-f_{i-1})+(f_{i}-f_{i+1})}$$
*Donde*: $L_{i}: Límite\ de\ la\ clase\ modal$\
$f_{i}:Frecuencia\ absulta$\
$F_{i-1}:Frecuencia\ absoluta \ inferior\ a\ la\ clase\ modal$\
$F_{i+1}:Frecuencia\ absoluta \ posterior\ a\ la\ clase\ modal$\
$a_{i}:Amplitud\ de\ la\ clase$


```python
qualification  = [9, 10, 8, 7, 6, 10, 4, 6, 9, 7, 8, 6, 9, 5, 6 ] #conjunto de datos 
print(stats.mode(qualification )) # se imprime el valor de la moda o promedio implementada  por la libreria statitics 
```

***NOTA***

***Población:*** Conjunto de **todos los elementos** cuyas caracteristicas se van a estudiar .

***Muestra:*** Subconjunto de la población.

**Varianza:** La varianza como medida de dispersión representa la variabilidad de una serie de datos respecto a su media. Formalmente se calcula como la suma de los residuos al cuadrado divididos entre el total de observaciones.
También se puede calcular como la desviación típica al cuadrado. 
Exiten dos maneras de calcular la varianza, ya sea de una muestra $s^2$ o de una población $\sigma^2$

$$S^2=\frac{\sum_{i=1}^n (x_{i}-\bar{x})^2}{n-1}$$\
$\bar{x}: es\ la\ media$\
$n:tamaño\ de\ la\ muestra$

```python
#varianza población: 
ppesos = [50, 60, 90, 39, 59, 70, 38, 40, 41] #conjunto de datos 
print(stats.pvariance(ppesos))  # se imprime el valor de la varianza tomando una población  implementada  por la libreria statitics 
```

```python
#varianza muestra: 
pesos = [57, 39, 63, 80, 34, 70, 40] #conjunto de datos 
print(stats.variance(pesos)) # se imprime el valor de la varianza tomando una muestra implementada  por la libreria statitics 
 
```

**Desviación estandar:** La desviación estándar mide la dispersión de una distribución de datos. Entre más dispersa está una distribución de datos, más grande es su desviación estándar.

$$\sigma=\sqrt{\frac{\sum(x-\bar{x})^2}{N-1}}$$\
Es interesante que la desviación estándar no puede ser negativa. Una desviación estándar cercana a 0 indica que los datos tienden a estar más cerca a la media . Entre más lejos estén los datos de la media, más grande es la desviación estándar.


```python
#Ejemplo con datos d elos pesos de personas
#muestra
pesos = [57, 39, 63, 80, 34, 70, 40] #conjunto de datos 
print(stats.stdev(pesos)) # se imprime el valor de la desviación estandar tomando una muestra implementada  por la libreria statitics 
 
```

```python
#población
ppesos = [50, 60, 90, 39, 59, 70, 38, 40, 41]  # se imprime el valor de la desviación estandar tomando una población implementada  por la libreria statitics 
 
print(stats.pstdev(ppesos))  
```
